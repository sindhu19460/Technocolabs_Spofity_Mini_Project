# Imported necessary lib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
session_data = pd.read_csv("log_mini.csv") #Import the data
```

```python
session_data.head(3)
```

```python
session_data.info() #information f the data
```

```python
col = ['skip_1','skip_2','skip_3',
       'not_skipped','hist_user_behavior_is_shuffle','premium']
```

```python
for colname in col:
    session_data[colname] = session_data[colname].astype(int) #was converting bool to int
```

```python
session_data.info() #data info
```

```python
session_data.rename(columns={'track_id_clean':'track_id'}, inplace=True) #renaming the track_id_clean to track_id
```

```python
session_data['skipped'] = session_data['skip_1']*session_data['skip_2']*session_data['skip_3'] 
#'''I think 3 columns inot have a necessay so i was multpliging those 3 skip_1,2,3'''
```

```python
session_data.drop(['not_skipped','skip_1','skip_2','skip_3'],axis=1,inplace=True) #No was droping those 4 cloumns
```

```python
session_data.info() #session_data info
```

```python
session_data["track_index"]=session_data["session_position"]
```

# Importing the track_feature data

```python
track_features = pd.read_csv("tf_mini.csv")
```

# Merging the data

```python
merged_data = session_data.merge(track_features, left_on="track_id",right_on="track_id",how="left") 
#'''was merging the track and train '''
```

```python
merged_data.drop("track_id",axis=1,inplace=True) #was doing track_id
```

```python
categorical_columns = ['context_type','hist_user_behavior_reason_start','hist_user_behavior_reason_end']
```

# One Hot Encoder

```python
import category_encoders as ce
```

```python
one_hot_encoder= ce.OneHotEncoder(cols=categorical_columns,handle_unknown=0,return_df=True,use_cat_names=True)
```

```python
merged_data_one_hot = one_hot_encoder.fit_transform(merged_data)
merged_data_one_hot.head()
```

# Data spliting

```python
y = merged_data_one_hot['skipped']
```

```python
merged_data_one_hot.drop(['skipped'],axis=1,inplace=True)
```

```python
merged_data_one_hot.shape
```

```python
merged_data_one_hot.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(merged_data_one_hot,y,test_size=0.3,random_state = 42)
```

```python
print(X_train.shape, y_train.shape)
```

```python
print(X_test.shape, y_test.shape)
```

# 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF = RandomForestClassifier(n_estimators = 120,
                           random_state = 50,
                           n_jobs = -1,
                           max_features = 'auto')
RF.fit(X_train,y_train)
```

```python
pred = RF.predict(X_test)
```

```python
from sklearn.metrics import recall_score,f1_score,roc_auc_score,precision_score,classification_report, confusion_matrix
```

```python
print("F1 Score",f1_score(pred,y_test))
```

```python
print("precision_score",precision_score(pred,y_test))
```

```python
print("roc_auc_score",roc_auc_score(pred,y_test))
```

```python
print("classification\n\n",classification_report(pred,y_test))
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd = SGDClassifier(alpha = 0.00047, random_state = 50)
```

```python
sgd.fit(X_train, y_train)
```

```python
sgd_pred = sgd.predict(X_test)
```

```python
print("Classification of SGD:\n\n", classification_report(y_test,sgd_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression(solver = 'liblinear',
                       multi_class = 'ovr',
                       max_iter = 1000,
                       random_state = 42,
                       penalty ="l2")
LR.fit(X_train,y_train)
```

```python
LR_pred = LR.predict(X_test)
```

```python
print("Classification of Logistic Regression\n\n",classification_report(y_test,LR_pred))
```

```python
import xgboost as xg
```

```python
xgb = xg.XGBClassifier(learning_rate = 0.01,
                       colsample_bytree = 0.8,
                       subsample = 0.8,
                       objective = 'multi:softmax', 
                       n_estimators = 100, 
                       reg_alpha = 0.3,
                       max_depth = 4, 
                       gamma = 1,
                       num_class = 3)
```

```python
xgb.fit(X_train , y_train);
```

```python
xgb_pred = xgb.predict(X_test)
```

```python
print("Classification of XGB\n\n",classification_report(y_test,xgb_pred),"\n")
```
